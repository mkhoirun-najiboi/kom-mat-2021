# Metode Regressi

metode yang saya pilih untuk menganalisis data 4 adalah metode `Linear Regression`. 
Linear Regression adalah sebuah pendekatan untuk memodelkan hubungan antara variabel terikat y dan satu atau lebih variabel bebas yang disebut X.

Disini saya akan membandingkan manakah yang lebih baik antara pengaruh Temperature terhadap Relative Humidity dan pengaruh Precipitation Total terhadap Relative Humidity.

Disini variabel terikat (y) adalah Relative Humidity (Kelembaban), dan variabel bebas (X) adalah Temperature (Suhu) dan Precipitation Total (Curah Hujan).

Package yang saya gunakan adalah `Statistics`.

```julia
using Statistics
```

```julia
X=dataset[2:end,2] #Data Temperature
y=dataset[2:end,3] #Data Relative Humidity
z=dataset[2:end,4] #Data Precipitation Total
```

```julia
num_tr_ex=length(y) 
```

```julia
y1=scatter(X,y, color="blue", xaxis="Temperature", markersize=5, yaxis="Relative Humidity")
```

Plot di atas adalah penyebaran dari data Temperature dan Relative humidity. 

```julia
y2=scatter(X,z, color="black", xaxis="Temperature", markersize=5, yaxis="Precipitation Total")
```

Plot di atas adalah penyebaran dari data Temperature dan Precipitation Total. 

```julia
using GLM
```

Pada table dibawah terdapat tanda “~” maksudnya yaitu nilai Y  dipengaruhi oleh nilai X, dimakalah ini Y dilambangkan dengan Q dan X yang dilambangkan dengan P.

```julia
Newdata=DataFrame(P=temp,Q=rh)
linearRegression1 = lm(@formula(Q ~ P), Newdata)
```

Pada table dibawah terdapat tanda “~” maksudnya yaitu nilai Y  dipengaruhi oleh nilai X, dimakalah ini Y dilambangkan dengan Q dan X yang dilambangkan dengan R.

```julia
Newdata=DataFrame(Q=rh,R=pt)
linearRegression2 = lm(@formula(Q ~ R), Newdata)
```

Kita akan mencari R square, yaitu untuk melihat seberapa besar variabel dependen itu dapat menjelaskan model.

```julia
r2(linearRegression1) #Temperature terhadap Relative Humidity
```

```julia
r2(linearRegression2) #Precipitation Total terhadap Relative Humidity
```

```julia
linearFit=predict(linearRegression1)
plot(temp,linearFit)
scatter!(temp,rh, color="blue", markersize=5, xaxis="Temperature", yaxis="Relative Humidity")
```

```julia
linearFit=predict(linearRegression2)
plot(pt,linearFit)
scatter!(pt,rh, color="black", markersize=5, xaxis="Temperature", yaxis="Precipitation Total")
```

Selanjutnya, kita akan mencari korelasi, di sini saya menggunakan `korelasi pearson`.

Korelasi pearson merupakan salah satu ukuran korelasi yang digunakan untuk mengukur kekuatan dan arah hubungan linier dari dua variabel.

Untuk korelasi ini, saya akan mencari korelasi antara Temperature dan Relative Humidity serta korelasi antara Precipitation Total dan Relative Humidity

0,00 - 0,199 : sangat rendah,
0,20 - 0,399 : rendah,
0,40 - 0,599 : sedang,
0,60 - 0,799 : kuat,
0,80 - 1,000 : sangat kuat,
Berdasarkan pedoman di atas data yang semakin mendekati 1 maka data tersebut semakin valid.

```julia
Statistics.cor(temp,rh)
```

```julia
Statistics.cor(pt,rh)
```

Korelasi pearson antara variabel Temperature dan Relative Humidity sebesar 0,5106. Karena nilai korelasi berada di range 0,40 - 0,599 maka disimpulkan bahwa hubungan antara kedua variabel tersebut sedang.

Korelasi pearson antara variabel Precipitation Total dan Relative Humidity sebesar 0,8033. Karena nilai korelasi berada di range 0,80 - 1,000 maka disimpulkan bahwa hubungan antara kedua variabel tersebut sangat kuat.
