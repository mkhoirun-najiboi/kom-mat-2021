# 4. Analisis dan Visualisasi Data Menggunakan Metode PCA

4.1 Analisis data menggunakan PCA

```julia
#Hasil proses smoothing dari setiap variabel digabungkan menjadi satu
DN=[z1 z2 z3 z4 z5]
```

```julia
using Statistics
dataN=DN
dataN = (dataN .- mean(dataN,dims = 1))./ std(dataN,dims=1) #Standardisasi data
```

```julia
using StatsBase
using MultivariateStats
dataN'
p = fit(PCA,dataN',maxoutdim=2) #Dari 5 variabel/dimensi, akan diekstraksi menjadi 2 dimensi/PC
```

```julia
M = projection(p) #Membentuk matriks proyeksi
```

```julia
principalvars(p) #Kontribus setiap PC dalam menjelaskan total varians
```

```julia
principalvars(p) ./ tvar(p) * 100 #Persentase total varians yang dijelaskan okeh tiap PC
```

4.2 Visualisasi hasil PCA

```julia
pcVar=principalvars(p)./ tvar(p)
cumVar=cumsum(pcVar)
pcDat=M[:,1:2]'*dataN'
```

```julia
pl1=plot(pcVar, c=:orange, label="Variance due to PC")
plot!(1:length(cumVar), cumVar, label="Cumulative Variance", c=:green, xlabel="Principal Component", ylabel="Variance Proportion", ylims=(0,1.5))
pl2=scatter(pcDat[1,:],pcDat[2,:], c=:red, xlabel="PC 1", ylabel="PC2", msv=0, legend=:none)
plot(pl1,pl2,size=(700,400))
```

```julia
pl1 |> save("Plot Varians.png")
pl2 |> save("Plot PC1 PC2.png")
```

```julia
q1=plot(x,z1, linewidth = 2, xlabel="Waktu", ylabel="Temperature", legend=false)
q2=plot(x,z2, linewidth = 2, xlabel="Waktu", ylabel="Precipitation Total", legend=false)
q3=plot(x,z3, linewidth = 2, xlabel="Waktu", ylabel="Relative Humidity", legend=false)
q4=plot(x,z4, linewidth = 2, xlabel="Waktu", ylabel="Wind Speed", legend=false)
q5=plot(x,z5, linewidth = 2, xlabel="Waktu", ylabel="Wind Direction", legend=false)
q6=plot(x,pcDat[1,:], linewidth = 2, label="PC 1", color=:red)
q7=plot(x,pcDat[2,:], linewidth = 2, label="PC 2", color=:orange)
plot(q1, q2, q3, q4, q5, q6, q7, size=(1300,1100))
```

```julia
compp=plot(q1, q4, q6, q7, size=(900,700))
compp |> save("Comp Plot.png")
```

4.3 Menghitung Norm

```julia
Xr = reconstruct(p, pcDat) #Pembentukan ulang matriks pendekatan terhadap matriks dataN' berdasarkan principal component
```

```julia
using LinearAlgebra
norm(Xr-dataN') #Menghitung error antara matriks Xr dengan dataN'
```
