# Metode ARIMA

Pada tugas ini, akan dianalisa dan ditentukan model peramalan dari metode ARIMA yang cocok untuk meramalkan data inflasi pada salah satu IHK yaitu sandang. Metode ARIMA juga dikenal sebagai metode runtut waktu Box-Jenkins. Metode ini merupakan gabungan antara model Autoregressive (AR) dan Moving Average (MA) yang dikembangkan oleh George Box dan Gwilyn Jenkins (Wulandari et al, 2019).

#### **Menentukan parameter AR dan MA dengan analisis plot ACF dan PACF**

ACF merupakan singkatan dari Autocorrelation Function, yang mengindikasikan nilai autoregressive. PACF merupkan singkatan Parcial Autocorrelation Function, yang mengindikasikan nilai Moving Averagenya. Dalam menentukan data mengandung Autoregressive (AR) atau Moving Average (MA) adalah melihat pola atau perilaku ACF dan PACFnya.

```julia
#Plot ACF dan PACF
total_lags = 15
s1 = scatter(collect(1:total_lags), autocor(d_train, collect(1:total_lags)), title = "ACF", ylim = [-0.3,0.5])
s2 = scatter(collect(1:total_lags), pacf(d_train, collect(1:total_lags)), title = "PACF", ylim = [-0.3,0.5])
plot(s1, s2, layout = (2, 1))
```

Dilihat dari gambar 6 dan gambar 7, terlihat bahwa kandidat model untuk peramalan yaitu:
Berdasarkan plot ACF -> ARIMA (0,1,1), Berdasarkan plot PACF -> ARIMA (1,1,0)


Agar penentuan model terbaik memiliki tingkat kepercayaan tinggi, kita uji dua model ARIMA yaitu ARIMA (1, 1, 0) dan ARIMA (0, 1, 1). Model ARIMA (1,1,0) didefinisikan sebagai berikut.

#### **Pengujian Residu**

```julia

@model ARIMA110(x) = begin
    T = length(x)
    μ ~ Uniform(-10, 10)
    ϕ ~ Uniform(-1, 1)
    for t in 3:T
        val = μ +                      # Drift term.
              x[t-1] +                 # bagian ARIMA(0,1,0).
              ϕ * (x[t-1] - x[t-2])    # bagian ARIMA(1,0,0).
        x[t] ~ Normal(val, 1)
    end
end
```

```julia
chain_ARIMA110 = sample(ARIMA110(d_train), NUTS(200, 0.65), 1500 )
```

```julia
plot(chain_ARIMA110, title = "Sampler plot for ARIMA(1, 1, 0) model")
```

Dari plot di atas, terlihat bahwa residu white noise dan berdistibusi normal

```julia
@model ARIMA011(x) = begin
    T = length(x)

    # Mendefinisikan error vector.
    ϵ = Vector(undef, T)
    x_hat = Vector(undef, T)

    θ ~ Uniform(-5, 5)

    # Treat the first x_hat as a parameter to estimate.
    x_hat[1] ~ Normal(550, 220)
    ϵ[1] = x[1] - x_hat[1]

    for t in 2:T
        # nilai prediksi untuk x.
        x_hat[t] = x[t-1] - θ * ϵ[t-1]
        # Menghitung observed error.
        ϵ[t] = x[t] - x_hat[t]
        # Mengobservasi likelihood.
        x[t] ~ Normal(x_hat[t], 1)
    end
end
```

```julia
chain_ARIMA011 = sample(ARIMA011(d_train), NUTS(200, 0.65), 10000 )
```

```julia
plot(chain_ARIMA011, title = "Sampler plot for ARIMA(0, 1, 1) model")
```

Dari plot di atas, terlihat bahwa residu white noise dan berdistibusi normal

#### **Melihat keakuratan Model Peramalan dengan nilai AIC**

```julia

# Menghitung nilai AIC dari ARIMA 110

function calculate_aic_ARIMA110(β::Float64, μ:: Float64, σ::Float64, d::Array{Float64, 1})
    T = length(d)
    ϵ = Vector(undef, T)
    d_pred = Vector(undef, T)
    
    d_pred[1], d_pred[2] = d[1], d[2]
    ϵ[1], ϵ[2] = 0.0, 0.0 
    for t in 3:T
        d_pred[t] = μ +                      
              d_pred[t-1] +                 
              β * (d_pred[t-1] - d_pred[t-2]) 
        ϵ[t] = d_pred[t] - d[t]
    end
    log_likelihood = (-(T - 1)/2)*2*π*σ^2 - (1/σ^2)*sum(ϵ.^2) - π*σ^2/(1 - β^2) - ((d[1] - μ/(1 - β))^2)/(2*σ^2/(1 - β^2))
    aic = -2*log_likelihood + 2
    return aic
end
```

```julia
calculate_aic_ARIMA110(mean(chain_ARIMA110[:ϕ]), mean(chain_ARIMA110[:μ]), 1.0, d_train)
```

```julia
# Menghitung nilai AIC dari ARIMA 011 
function calculate_aic_ARIMA011(β::Float64, σ::Float64, d::Array{Float64, 1})
    T = length(d)

    ϵ = Vector(undef, T)
    d_pred = Vector(undef, T)

    d_pred[1] = d[1]
    ϵ[1] = 0.0
    for t in 2:T
        d_pred[t] = d[t-1] - β * ϵ[t-1]
        ϵ[t] = d[t] - d_pred[t]
    end
    log_likelihood = -(T/2)*log(2*π) - (T/2)*log(σ^2) + (1/2*σ^2)*sum(ϵ.^2)  
    aic = -2*log_likelihood + 1
    return aic
end
```

```julia
calculate_aic_ARIMA011(mean(chain_ARIMA011[:θ]), 1.0, d_train)
```

Semakin kecil nilai AIC, maka model yang dipilih semakin akurat.

Berdasarkan nilai AIC, model terbaik yang tepat untuk meramalkan data inflasi sandang adalah ARIMA (0,1,1)
