#  Harga Saham

```julia
using Plots
using StatsBase
using DataFrames
using XLSX
using StatsPlots
```

```julia
A = XLSX.readxlsx("Data 1.xlsx")
```

```julia
Harga_Saham = A["Harga Saham"][:]
Harga_Saham = DataFrame(Harga_Saham[2:end,:],Symbol.(Harga_Saham[1,:]))
```

```julia
Data_HS = Matrix(Harga_Saham)
B =Float64.(Data_HS[1:end,2:end])
```

```julia
Data_HS = Matrix(Harga_Saham)
C =Float64.(Data_HS[:,2: 3])
```

```julia
Data_HS = Matrix(Harga_Saham)
D =Float64.(Data_HS[:,4: 5])
```

```julia
times = Data_HS[1:end,:1]
```

## ðŸŸ£ Plot/Scatter semua data harga saham

```julia
C1 = times
C2 = C
scatter(C1,C2,size=(1000,500), title="Harga Saham WIKA - BISI", legend =false, labels=["WIKA" "BISI"], xlabel=("Bulan_Tahun"), ylabel=("Harga"))
plot!(C1,C2)
```

```julia
D1 = times
D2 = D
scatter(D1,D2,size=(1000,500), title="Harga Saham BSDE - NIKL", legend =false, color=["dark blue" "pink"], labels=["BSDE" "NIKL"],  xlabel=("Bulan_Tahun"), ylabel=("Harga"))
plot!(D1,D2,)
```

## ðŸŸ£ Plot with Boxplot Data Harga Saham

```julia
boxplot(B,  bins=50, alpha=0.5, labels=["WIKA" "BISI" "BSDE" "NIKL"])
```

## ðŸŸ£ Menentukan mean, median, kuartil, min, dan max dari data harga saham

```julia
println("Deskripsi statistik Harga Saham WIKA : ")
describe(B[: ,1])
```

```julia
println("Deskripsi statistik Harga Saham BISI: ")
describe(B[: ,2])
```

```julia
println("Deskripsi statistik Harga Saham BSDE: ")
describe(B[: ,3])
```

```julia
println("Deskripsi statistik Harga Saham NIKL: ")
describe(B[: ,4])
```
