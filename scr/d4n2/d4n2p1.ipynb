# Scatter Plot (Data per Hari)

```julia
per_hari = DataFrame(XLSX.readtable("Data_4.xlsx","Per Hari")...)
```

```julia
using Plots
```

```julia
urutan = per_hari[1:end,1]
```

```julia
scatter(urutan,per_hari[1:end,3],legend=false)
xlabel!("urutan waktu")
ylabel!("rata-rata suhu di Jakarta (°C)")
```

```julia
scatter(urutan,per_hari[1:end,4],legend=false)
xlabel!("urutan waktu")
ylabel!("total curah hujan di Jakarta (mm)")
```

```julia
scatter(urutan,per_hari[1:end,5],legend=false)
xlabel!("urutan waktu")
ylabel!("rata-rata kelembaban di Jakarta (%)")
```

```julia
scatter(urutan,per_hari[1:end,6],legend=false)
xlabel!("urutan waktu")
ylabel!("rata-rata kecepatan angin di Jakarta (km/h)")
```

```julia
scatter(urutan,per_hari[1:end,7],legend=false)
xlabel!("urutan waktu")
ylabel!("rata-rata arah angin di Jakarta (°)")
```

# Violin Plot (Data per Bulan)

```julia
using Plots
using StatsPlots
gr()
```

```julia
using Statistics
using StatsBase
using MLBase
```

```julia
per_bulan = DataFrame(XLSX.readtable("Data_4.xlsx","Per Bulan")...)
```

```julia
violin(["Suhu"], per_bulan[1:end,3],legend=false,alpha=0.8)
violin!(["Curah Hujan"], per_bulan[1:end,4],legend=false,alpha=0.8)
violin!(["Kelembaban"], per_bulan[1:end,5],legend=false,alpha=0.8)
violin!(["Kecepatan Angin"], per_bulan[1:end,6],legend=false,alpha=0.8)
violin!(["Arah Angin"], per_bulan[1:end,7],legend=false,alpha=0.8)
```

```julia
scatter(per_bulan[1:end,1],per_bulan[1:end,4],legend=false)
xlabel!("Bulan Ke-")
ylabel!("Total curah hujan (mm)")
title!("Total Curah Hujan di Jakarta Tahun 2019")
```
