```julia
samarinda=datamcurah[1:120,1]
#nilai cs Samarinda
ns=length(samarinda)
meansamarinda = mean(samarinda)
xsamarinda=datamcurah[1:120,1]
ssamarinda=datamcurah[1:120,1]
#nilai sigma
for i in 1:ns
   xsamarinda[i]=(samarinda[i]-meansamarinda).^3
end
#nilai standar deviasi
for i in 1:ns
    ssamarinda[i]=(samarinda[i]-meansamarinda).^2
end
s=sum(ssamarinda)
stdsamarinda=sqrt(s/(ns-1))
a=(ns/((ns-1)*(ns-2)))*sum(xsamarinda)
cssamarinda=a/stdsamarinda.^3
```
